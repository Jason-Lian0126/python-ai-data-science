```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# 輸入股票代號
symbol = input("股票代號：")

# 時間範圍
start_date = "2022-01-01"
end_date = "2023-04-19"

# 使用yfinance獲取股票數據
stock_data = yf.download(symbol, start=start_date, end=end_date)

# 股價與交易量
price_data = stock_data["Close"]
volume_data = stock_data["Volume"]

# 繪圖製表
plt.figure(figsize=(12, 6))
plt.subplot(2, 1, 1)
plt.plot(stock_data.index, price_data)
plt.title(f"{symbol} Stock Prices")
plt.xlabel("Date")
plt.ylabel("Price")
plt.subplot(2, 1, 2)
plt.bar(stock_data.index, volume_data)
plt.title(f"{symbol} Volume")
plt.xlabel("Date")
plt.ylabel("Volume")
plt.tight_layout()
plt.show()
```

分析多個股價走勢

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt


symbol1 = input("請輸入第一個股票代碼：")
symbol2 = input("請輸入第二個股票代碼：")
symbol3 = input("請輸入第三個股票代碼：")

# 時間範圍
start_date = "2022-04-19"
end_date = "2023-04-19"

# 使用yfinance獲取股票數據
stock_data1 = yf.download(symbol1, start=start_date, end=end_date)
stock_data2 = yf.download(symbol2, start=start_date, end=end_date)
stock_data3 = yf.download(symbol3, start=start_date, end=end_date)

# 股價與交易量
price_data1 = stock_data1["Close"]
volume_data1 = stock_data1["Volume"]
price_data2 = stock_data2["Close"]
volume_data2 = stock_data2["Volume"]
price_data3 = stock_data3["Close"]
volume_data3 = stock_data3["Volume"]

# 繪製圖表
plt.figure(figsize=(12, 6))
plt.subplot(2, 1, 1)
plt.plot(stock_data1.index, price_data1, label=symbol1)
plt.plot(stock_data2.index, price_data2, label=symbol2)
plt.plot(stock_data3.index, price_data3, label=symbol3)
plt.title("Stock Prices")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.subplot(2, 1, 2)
plt.bar(stock_data1.index, volume_data1, label=symbol1)
plt.bar(stock_data2.index, volume_data2, label=symbol2)
plt.bar(stock_data3.index, volume_data3, label=symbol3)
plt.title("Volume")
plt.xlabel("Date")
plt.ylabel("Volume")
plt.legend()
plt.tight_layout()
plt.show()
```
