```python
import yfinance as yf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM

# Use yfinance to get the historical stock price data
tsmc = yf.Ticker("2330.TW").history(period="1y")

# Scale the stock price data to the range of 0 to 1
scaler = MinMaxScaler(feature_range=(0, 1))
tsmc_scaled = scaler.fit_transform(tsmc.Close.values.reshape(-1, 1))

# Define the number of timesteps and the number of features
timesteps = 30
num_features = 1

# Define the input data and the target data
X = []
y = []
for i in range(timesteps, tsmc_scaled.shape[0]):
    X.append(tsmc_scaled[i-timesteps:i, 0])
    y.append(tsmc_scaled[i, 0])
X, y = np.array(X), np.array(y)
X = np.reshape(X, (X.shape[0], X.shape[1], num_features))

# Define the RNN model
model = Sequential()
model.add(LSTM(50, return_sequences=True, input_shape=(X.shape[1], num_features)))
model.add(LSTM(50))
model.add(Dense(1))
model.compile(loss='mean_squared_error', optimizer='adam')

# Fit the RNN model to the stock price data
model.fit(X, y, epochs=100, batch_size=32)

# Predict the future stock price for the next 7 days
inputs = tsmc_scaled[-timesteps:]
inputs = np.reshape(inputs, (1, timesteps, num_features))
prediction = model.predict(inputs)
prediction = scaler.inverse_transform(prediction)
```
